```python
geolocater = Nominatim(user_agent = "mapping store locations to county")
```

```python
cabelas = pd.read_csv("data/cabelas_location.csv", header=None).rename(columns={0: "store_location"})
cinemarks = pd.read_csv("data/cinemark_location.csv", header=None).rename(columns={0: "store_location"})
nordstroms = pd.read_csv("data/nordstrom_location.csv", header=None).rename(columns={0: "store_location"})
trader_joes = pd.read_csv("data/trader_joes_location.csv", header=None).rename(columns={0: "store_location"})
```

```python
def get_county(df, store_name):
    counties = []
    for i in range(len(df['store_location'])):
        address = df['store_location'][i]
        if (store_name =='nordstroms' or store_name=='trader_joes') and ('10003' in address or '01803' in address):
            counties.append(None)
            continue
        loc = geolocater.geocode(address)
        found_space = True 
        if not loc:
            if " " in address:
                address = address.split(" ", 1)[1]
                loc = geolocater.geocode(address)
        while not loc:
            if ", " in address:
                address = address.split(" ", 1)[1]
                loc = geolocater.geocode(address)
    #             if not loc:
    #                 print(address)
            else:
                found_space = False
                break
        if found_space:
            loc_sections = loc.address.split(", ")
            try:
                counties.append([l for l in loc_sections if ((" County" in l) or (" Parish" in l))][0])
            except:
                counties.append(None)
                print(loc_sections)
        if not found_space:
            counties.append(None)
        
    df[store_name + "_county"] = counties
```

```python
# https://gist.github.com/rogerallen/1583593
us_state_abbrev = {
    'Alabama': 'AL',
    'Alaska': 'AK',
    'American Samoa': 'AS',
    'Arizona': 'AZ',
    'Arkansas': 'AR',
    'California': 'CA',
    'Colorado': 'CO',
    'Connecticut': 'CT',
    'Delaware': 'DE',
    'District of Columbia': 'DC',
    'Florida': 'FL',
    'Georgia': 'GA',
    'Guam': 'GU',
    'Hawaii': 'HI',
    'Idaho': 'ID',
    'Illinois': 'IL',
    'Indiana': 'IN',
    'Iowa': 'IA',
    'Kansas': 'KS',
    'Kentucky': 'KY',
    'Louisiana': 'LA',
    'Maine': 'ME',
    'Maryland': 'MD',
    'Massachusetts': 'MA',
    'Michigan': 'MI',
    'Minnesota': 'MN',
    'Mississippi': 'MS',
    'Missouri': 'MO',
    'Montana': 'MT',
    'Nebraska': 'NE',
    'Nevada': 'NV',
    'New Hampshire': 'NH',
    'New Jersey': 'NJ',
    'New Mexico': 'NM',
    'New York': 'NY',
    'North Carolina': 'NC',
    'North Dakota': 'ND',
    'Northern Mariana Islands':'MP',
    'Ohio': 'OH',
    'Oklahoma': 'OK',
    'Oregon': 'OR',
    'Pennsylvania': 'PA',
    'Puerto Rico': 'PR',
    'Rhode Island': 'RI',
    'South Carolina': 'SC',
    'South Dakota': 'SD',
    'Tennessee': 'TN',
    'Texas': 'TX',
    'Utah': 'UT',
    'Vermont': 'VT',
    'Virgin Islands': 'VI',
    'Virginia': 'VA',
    'Washington': 'WA',
    'West Virginia': 'WV',
    'Wisconsin': 'WI',
    'Wyoming': 'WY'
}
# thank you to @kinghelix and @trevormarburger for this idea
abbrev_us_state = dict(map(reversed, us_state_abbrev.items()))


def get_state(df, store_name):
    states = []
    for i in range(len(df['store_location'])):
        address = df['store_location'][i]
        for abbrev in abbrev_us_state:
            if abbrev in address:
                states.append(abbrev_us_state[abbrev])
                break
        else:
            states.append(None)
            print(store_name)
            print(address)
        
    df[store_name + "_state"] = states
```
